```python
from langsmith import traceable
import random
import time

@traceable
def gen_rand_number():
  return random.randint(0,100)


@traceable
def gen_str_delay(input_str:str):
  number=random.randint(1,5)
  time.sleep(number)
  return f"{input_str}({number})"
```

```python
from tqdm.auto import tqdm

for _ in tqdm(range(10)):
    gen_rand_number()
    gen_str_delay("Hello")
```
